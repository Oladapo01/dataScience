```python
import pandas as pd 

dataset = 'COM618 dataset.csv'
data = pd.read_csv(dataset)

print(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Breakdown Analysis

# 1. Gender and Diabetes
gender_diabetes = pd.crosstab(data['gender'], data['diabetes'], normalize='index') * 100

# 2. Age Groups and Smoking Status
# Creating age groups
bins = [0, 18, 30, 40, 50, 60, 70, 80]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71-80']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

# Analyzing smoking status across different age groups
smoking_age_group = pd.crosstab(data['age_group'], data['smoking_history'], normalize='index') * 100

# 3. Smoking Status and Diabetes
smoking_diabetes = pd.crosstab(data['smoking_history'], data['diabetes'], normalize='index') * 100

# Visualizing the results
plt.figure(figsize=(14, 10))

# Gender and Diabetes
plt.subplot(3, 1, 1)
sns.barplot(x=gender_diabetes.index, y=gender_diabetes[1])
plt.title('Percentage of Diabetes by Gender')
plt.ylabel('Percentage with Diabetes')

# Age Groups and Smoking Status
plt.subplot(3, 1, 2)
smoking_age_group.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Smoking History Across Different Age Groups')
plt.ylabel('Percentage')

# Smoking Status and Diabetes
plt.subplot(3, 1, 3)
sns.barplot(x=smoking_diabetes.index, y=smoking_diabetes[1])
plt.title('Percentage of Diabetes by Smoking History')
plt.ylabel('Percentage with Diabetes')

plt.tight_layout()
plt.show()

gender_diabetes, smoking_age_group, smoking_diabetes
```

```python

# Checking data types and missing values
data_info = data.info()
missing_values = data.isnull().sum()

# Statistical summary of numerical columns
statistical_summary = data.describe()

# Visualizations of numerical variables
plt.figure(figsize=(15, 10))

# Histograms for numerical variables
for i, col in enumerate(['age', 'smoking_history', 'gender'], start=1):
    plt.subplot(2, 2, i)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()

# Preparing data for categorical variable analysis
categorical_columns = ['gender','age', 'smoking_history']
categorical_data = data[categorical_columns]

# Visualizations of categorical variables
plt.figure(figsize=(15, 10))

for i, col in enumerate(categorical_columns, start=1):
    plt.subplot(3, 2, i)
    sns.countplot(x=col, data=categorical_data)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()

# Output: Data info, missing values, and statistical summary
data_info, missing_values, statistical_summary
```
